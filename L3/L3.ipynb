# Задание 1.

## Ключевой процесс - Реализация строительного объекта



*   Цель процесса:
Построить объект строительства в соответствии с проектной документацией, сроками и бюджетом.

*   Инициатор:
Руководство организации



*   Выходные данные:
Выполненные этапы работ.
Акты приемки выполненных работ.
Исполнительная документация.


*   Участники:

    Начальник участка / Прораб.

    Инженерно-технический персонал.

    Рабочие / Бригады.

    Заказчик.



*   Ожидаемый результат:
Сданный в эксплуатацию объект (или его этап).








##Вспомогательный процесс №1 - закупка стротильеных матеарилов.



* Цепль процесса:
Обеспечить строительный объект необходимыми материалами и оборудованием в нужное время.

* Инициатор:
Начальник участка / Прораб (когда видит нехватку материалов).

* Входные данные:
Заявка на материалы.
Остатки на складе.

* Выходные данные:
Поставленные материалы на объект.
Товарные накладные и счета-фактуры.

* Участники:
    Начальник участка.
    Отдел снабжения / Закупщик.
    Поставщик.
    Бухгалтерия .

##Ожидаемый резульат:
Материалы находятся на стройплощадке, готовность к работе не простаивает.



##Вспомогательный процесс №2 - планирование работ и распределение ресурсов.



* Цепль процесса:
Составить реалистичный график работ и назначить на объекты нужные бригады и технику.

* Инициатор:
Главный инженер.

* Входные данные:
Условия договора.
Нормы выработки.
Информация о доступности техники и людей.

* Выходные данные:
Календарный график работ.
Наряды на работу.
Путевые листы для техники.

* Участники:
    Планово-экономический отдел.
    Руководство.
    Начальники участков.

##Ожидаемый результат:

Утвержденный план, по которому работают все подразделения.
