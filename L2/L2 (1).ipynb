# Анализ стейкхолдеров

**1. Руководство организации**

Интерес: повышение эффективности управления, прозрачность процессов

Цели:
контроль сроков строительства
снижение затрат
получение аналитических отчетов

Влияние: высокое

Риски:
изменение требований в процессе разработки
 и завышенные ожидания

**2. Начальники строительных управлений**

Интерес: контроль подчинённых подразделений

Цели:
управление участками
контроль выполнения работ

Влияние: высокое

Риски:
сопротивление изменениям
и неполное использование системы

**3. Начальники участков**

Интерес: удобный инструмент управления работами

Цели:
планирование задач
контроль бригад
учет ресурсов

Влияние: среднее

Риски:
ошибки ввода данных и
игнорирование системы

**4. Инженерно-технический персонал**

Интерес: доступ к актуальной информации

Цели:
ведение учета работ
контроль выполнения задач

Влияние: среднее

Риски:
некорректные данные и
недостаток обучения

**5. Планово-экономический отдел**

Интерес: точные расчеты и планирование

Цели:
формирование смет
анализ затрат

Влияние: среднее

Риски:
ошибки в расчетах и
несоответствие факта и плана

**6. Бухгалтерия**

Интерес: корректный учет ресурсов и затрат

Цели:
учет материалов
финансовая отчетность

Влияние: среднее

Риски:
несовместимость с учетными системами и
ошибки интеграции

**7. Рабочие / Бригады**

Интерес: понятные задачи и распределение работ

Цели:
выполнение работ
получение заданий

Влияние: низкое

Риски:
отказ использовать систему
ошибки исполнения

**9. IT-специалисты / разработчики**

Интерес: успешная реализация проекта

Цели:
разработка системы
поддержка и сопровождение

Влияние: высокое

Риски:
ошибки проектирования и
нехватка ресурсо

# Границы системы

1. Что входит в систему

В границы системы входят следующие функции и процессы:

Управление организационной структурой
1) ведение строительных управлений
2) ведение участков
3) распределение сотрудников по подразделениям


Учет персонала и бригад
1) хранение данных о сотрудниках
2) учет должностей, квалификаций и категорий
3) формирование и ведение бригад
4) назначение бригадиров

Учет объектов строительства
1) ведение информации об объектах
2) хранение адреса, заказчика, типа объекта
3) учет состояния объекта

Учет видов работ
1) ведение справочника видов работ
2) привязка работ к объектам
3) контроль этапов выполнения

Учет материалов
1) ведение справочника материалов
2) учет планового количества материалов по смете
3) контроль использования материалов
4) выявление перерасхода

Учет строительной техники
1) ведение перечня техники
2) учет инвентарных номеров
3) распределение техники по объектам или участкам
4) контроль доступности техники

Формирование смет
1) создание смет по объектам
2) учет плановых объемов материалов
3) связь смет с объектами и работами

Планирование и контроль работ
1) составление графиков выполнения работ
2) контроль сроков
3) фиксация фактического состояния выполнения

Отчетность
1) формирование сводных отчетов
2) получение аналитической информации по объектам, материалам, срокам и технике

Администрирование системы
1) управление учетными записями пользователей
2) разграничение прав доступа
3) обеспечение сохранности данных

# Внешние системы


**Бухгалтерская система:**

Используется для передачи данных о затратах, материалах, финансовых показателях.


**Системы хранения документов:**

Могут использоваться для хранения договоров, актов, проектной документации и сопутствующих файлов.


# Какие процессы можно и нужно автоматизировать

**Нужно автоматизировать в первую очередь:**

1) учет объектов строительства
2) учет сотрудников и бригад
3) учет строительной техники
4) учет материалов
5) ведение смет
6) формирование графиков работ
7) контроль сроков выполнения
7) контроль перерасхода материалов
8) формирование отчетности
