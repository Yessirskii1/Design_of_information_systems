# Анализ предметной области
**Тема:**  Проектирование информационной системы строительной организации

**Название:** ProInstall

**Описание:** Сервис представляет собой информационную систему для автоматизации деятельности строительной организации. Система предназначена для учета объектов строительства , управления структурой организации , ведения учета персонала и бригад, распределения строительной техники, формирования графиков работ и смет, а также подготовки отчетов по результатам строительства.


# Предметная область

**Целевая аудитория:**

* Руководство строительной организации

* Начальники строительных управлений

* Начальники участков

* Инженерно-технический персонал

* Экономический и плановый отдел

* Сотрудники бухгалтерии

# Проблемы и текущее состояние

**Существующие проблемы и недостатки:**

Во многих строительных организациях учет ведется частично в таблицах Excel, частично в бумажном виде или в разрозненных программах. Это приводит к:
сложности контроля сроков;
трудностям отслеживания перерасхода материалов;
неэффективному распределению техники;
сложности формирования сводной отчетности;
риску потери данных.
Единая информационная система позволит централизовать управление и повысить прозрачность процессов.

# Цель и задачи проекта

**Цель:** Проектирование и разработка информационной системы для автоматизации управления строительной организацией.

**Задачи:**

* Провести анализ предметной области.

* Определить требования к системе.

* Спроектировать структуру базы данных.

* Реализовать учет объектов, персонала, техники и работ.

* Реализовать формирование графиков и смет.

* Реализовать формирование отчетов и аналитических запросов.

* Подготовить документацию

# Основные требования

**Функциональные требования:**

Система должна обеспечивать:

* ведение структуры организации;
* учет сотрудников и бригад;
* учет объектов строительства;
* формирование графиков выполнения работ;
* формирование смет;
* учет строительной техники;
* контроль сроков выполнения работ;
* контроль перерасхода материалов;


**Нефункциональные требования:**
* Минималистичный дизайн
* Наличие полноценных веб-версии и мобильной версии сервиса
* Хеширование паролей пользователей
* Хранение медиафайлов в S3 хранилище
* Сбор метрик работы сервиса
* Покрытие тестами
* Документация API
* Поддержка русского, английского
